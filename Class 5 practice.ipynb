```python
import pandas as pd
df['signup_date'] = pd.to_datetime(df['signup_date'])
```

```python
df['hour'] = df['signup_date'].dt.hour
```

```python
df['month'] = df['signup_date'].dt.month
```

```python
df['day'] = df['signup_date'].dt.day
```

```python
 def total_cat(x):
    if x >= 0 and x < 10:
        return '0-10'
    elif x >= 10 and x < 50:
        return '10-50'
    elif x >= 50 and x < 100:
        return '50-100'
    else:
        return '100+'
```

```python
df['total_purchases_group'] = df['total_purchases'].apply(total_cat)
```

```python
df.head()
```

```python
def good_bad(hour,day):
    if hour > 10 and day > 10:
        return 'good'
    elif hour <=10 and day >= 10 and day <= 20:
        return 'normal'
    else:
        return 'fine'
```

```python
df['user_type'] = df.apply(lambda x: good_bad(x['hour'], x['day']), axis=1)
```

```python
#Dummy variables  ---rename
gender_mapping = {1:'female', 2:'male'}
df['gender'] = df['gender'].map(gender_mapping)
```

```python
#Create gender dummy variables
gender_dummies = pd.get_dummies(df['gender'])
```
